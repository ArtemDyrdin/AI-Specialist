```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, roc_curve, auc
from sklearn.model_selection import train_test_split, GridSearchCV
```

```python
data = pd.read_csv(r"C:\Users\PAVILION\Desktop\Coding\Datasets\dataset_Facebook.csv", sep=';')
```

```python
data_copy = data
data_copy.head()
```

```python
data_copy.isnull().sum()[data_copy.isnull().sum() > 0]
```

```python
data_copy.info()
```

```python
data_copy[['like', 'share', 'Paid']] = data_copy[['like', 'share', 'Paid']].apply(pd.to_numeric, errors='coerce')
```

```python
data_copy['like'] = data_copy['like'].fillna(data_copy['like'].median())
data_copy['share'] = data_copy['share'].fillna(data_copy['share'].median())
```

```python
data_copy[data_copy['Paid'].isnull()].index
```

```python
data_copy = data_copy.drop(data_copy[data_copy['Paid'].isnull()].index)
```

```python
data_copy.info()
```

```python
X = data_copy.drop(['Paid'], axis=1)
X = pd.get_dummies(X)
```

```python
y = data_copy['Paid']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=1)
```

```python
logReg = LogisticRegression()
```

```python
logReg.fit(X_train, y_train)
```

```python
predicted = logReg.predict(X_test)
```

```python
confusion_matrix(y_test, predicted)
```

```python
predicted_proba = logReg.predict_proba(X_test)
```

```python
fpr, tpr, thresholds = roc_curve(y_test, predicted_proba[:, 1])
roc_auc = auc(fpr, tpr)
```

```python
plt.figure()
plt.plot(fpr, tpr, color='blue', label='ROC-кривая (area = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], color='red', linestyle='--')  # Линия случайного угадывания
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
```
