__Импортируем все библиотеки, которые будут использоваться__

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import roc_auc_score, roc_curve
import matplotlib.pyplot as plt
import seaborn as sns
```

__Загружаем данные__

```python
data = pd.read_csv('data/risk_factors_cervical_cancer.csv')
```

```python
data
```

__Заменим '?' на None, чтобы было проще обрабатывать пропуски__

```python
data.replace("?", np.nan, inplace=True)
```

Переводим 'object' в подходящий для на тип данных float

```python
for col in data.columns:
    if data[col].dtype == "object":
        try:
            data[col] = data[col].astype(float)
        except ValueError:
            continue
```

__Заполним пропуски медианой__

```python
data.fillna(data.median(numeric_only=True), inplace=True)
```

```python
data
```

__Разделим на признаки и целевую переменную__

```python
X = data.drop('Dx:Cancer', axis=1)
y = data['Dx:Cancer']
```

__Разделим на обучающую и тестовую выборки__

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)
```

Параметры для исследования глубины выборки:

```python
train_scores = []
test_scores = []
max_depths = range(1, 21)
```

__Построим дерево решений с использованием кросс-валидации__

```python
for depth in max_depths:
    # Создаем модель с заданной глубиной
    model = DecisionTreeClassifier(max_depth=depth, random_state=42)
    
    # 3-fold кросс-валидация на обучающей выборке
    train_cv_score = cross_val_score(model, X_train, y_train, cv=3, scoring="roc_auc").mean()
    train_scores.append(train_cv_score)
    
    # Обучение и оценка на тестовой выборке
    model.fit(X_train, y_train)
    test_pred = model.predict_proba(X_test)[:, 1]
    test_auc = roc_auc_score(y_test, test_pred)
    test_scores.append(test_auc)
```

__Построим графики__

```python
plt.figure(figsize=(10, 6))
plt.plot(max_depths, train_scores, label="Train ROC AUC", marker="o")
plt.plot(max_depths, test_scores, label="Test ROC AUC", marker="o")
plt.xlabel("Tree Depth")
plt.ylabel("ROC AUC")
plt.title("ROC AUC and Tree Depth")
plt.legend()
plt.grid()
plt.show()
```

__Построим финальную модель с оптимальной глубиной__

```python
optimal_depth = max_depths[np.argmax(test_scores)]
final_model = DecisionTreeClassifier(max_depth=optimal_depth, random_state=42)
final_model.fit(X_train, y_train)
```

__Покажем важность признаков__

```python
feature_importances = pd.Series(final_model.feature_importances_, index=X.columns)
feature_importances = feature_importances.sort_values(ascending=False)
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=feature_importances.values, y=feature_importances.index)
plt.title("Feature Importances")
plt.xlabel("Importance")
plt.ylabel("Features")
plt.show()
```

```python
print(f"Оптимальная глубина дерева: {optimal_depth}")
print(f"Лучший ROC AUC на тестовой выборке: {max(test_scores):.4f}")
print("Топ-5 наиболее важных признаков:")
print(feature_importances.head())
```

__Можно прийти к выводу, что DX:HPV напрямую однозначно определяет DX:Cancer, если все правильно сделали__

```python
intersection = np.logical_and(result_Cancer, result_hpv)
```

```python
np.count_nonzero(intersection==0)
```
